## Using SQLite3. Creating Table 1

### Table 1: Sno, MovieName, Director Name, Duration, metascore, ratings

```python
import sqlite3
import csv

# Connect to the SQLite database
db = sqlite3.connect("IMDB.db")

# Create a cursor object to interact with the database
cur = db.cursor()

# Create a table1 in the database. Used SNo as PRIMARY_KEY as there are unique numbers in this column and 
# used Movie_Name as FOREIGN KEY to create referential integrity between Table 1 and Table 2
cur.execute('''CREATE TABLE Table1 (
                    SNo INT PRIMARY KEY,
                    Movie_Name TEXT,
                    Duration_min INT,
                    Year INT,
                    Ratings REAL,
                    Metascore INT,
                    Director1 TEXT,
                    Director2 TEXT,
                    Director3 TEXT,
                    Director4 TEXT,
                    Director5 TEXT,
                    Director6 TEXT,
                    FOREIGN KEY (Movie_Name) REFERENCES Table2(Movie_Name)
                )''')

# Read data from the CSV file and insert into the table
with open('IMDB_1.csv', 'r', encoding='utf-8') as file: # Used encoding='utf-8' as default one gave issue
    csv_reader = csv.reader(file)  # Creating csv_reader object using the csv.reader() function. csv.reader reads the file 
                                   # and allows to iterate over its rows.
    next(csv_reader)  # Skip the header row if present
    
    total_records=0
    for row in csv_reader:
        # Insert each row of data into the table
        cur.execute("INSERT INTO Table1 VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)", row) # Dynamic insertion of data into 
                                                                                           # the table from csv file
        db.commit()  # Commit the changes to the database
        total_records+=1


print(total_records, "Records Inserted into Table1 from first CSV file")

# Close the database connection
db.close()
```

## Using SQLite3. Creating Table 2

### Table 2: Movie Name, stars, votes, Genre, Gross collection, popularity, Certification

```python
import sqlite3
import csv

# Connect to the SQLite database
db = sqlite3.connect("IMDB.db")

# Create a cursor object to interact with the database
cur = db.cursor()

# Create a Table2 in the database. Used "Movie_Name" as PRIMARY_KEY where as it is used as FOREIGN KEY in Table1.
cur.execute('''CREATE TABLE Table2 (
                    Movie_Name TEXT PRIMARY KEY,
                    Votes INT,
                    Certification TEXT,
                    Gross_Million_Dollars REAL,
                    Star1 TEXT,
                    Star2 TEXT,
                    Star3 TEXT,
                    Star4 TEXT,
                    Genre1 TEXT,
                    Genre2 TEXT,
                    Genre3 TEXT,
                    Popularity INT
                )''')

# Read data from the CSV file and insert into the table
with open('IMDB_2.csv', 'r', encoding='utf-8') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)  # Skip the header row if present
    
    total_records=0
    for row in csv_reader:
        # Insert each row of data into the table
        cur.execute("INSERT INTO Table2 VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)", row)
        db.commit()  # Commit the changes to the database
        total_records+=1


print(total_records, "Records Inserted into Table2 from second CSV file")

# Close the database connection
db.close()
```

## Make subquery using SQLite3: 

- **1) Display all details from both tables where movie names are the same.**

- **2) Display all movie names, Director, ratings, and gross collection where the genre is action.**

- **3) Display all details from both tables with the highest gross collection.**

- **4) Display all details from both tables with the highest ratings**

- **5) Display all details from both tables with the lowest gross collection and lowest ratings**

```python
import sqlite3

# dbect to the SQLite database
db = sqlite3.connect('IMDB.db')
cur = db.cursor()

# Subquery: Display all details from both tables where movie names are the same
subquery1 = """
    SELECT *
    FROM Table1
    INNER JOIN Table2 ON Table1.Movie_Name = Table2.Movie_Name
"""
cur.execute(subquery1)
results1 = cur.fetchall()
print("Subquery1: Display all details from both tables where movie names are the same")

# I am trying to print max 20 rows for display purpose. Displaying first 10 and last 10 rows for display purpose.

num_rows = len(results1)  # Determine the number of rows
max_rows = 20  # Determine the maximum number of rows to print

# Print the results
if num_rows > max_rows:
    # Print first 10 rows
    for row in results1[:10]:
        print(row)

    print("-"*100)
    print("-"*100)

    # Print last 10 rows
    for row in results1[-10:]:
        print(row)
else:
    # Print all results
    for row in results1:
        print(row)
```

```python
# Query: Display all movie names, Director, ratings, and gross collection where the genre is action
query2 = """
    SELECT Table2.Movie_Name, Table1.Director1, Table1.Director2, Table1.Director3, Table1.Director4, 
    Table1.Director5, Table1.Director6,  Table1.Ratings, Table2.Gross_Million_Dollars
    FROM Table1
    INNER JOIN Table2 ON Table1.Movie_Name = Table2.Movie_Name
    WHERE Table2.Genre1 = 'Action' OR Table2.Genre2 = 'Action' OR Table2.Genre3 = 'Action'
"""
cur.execute(query2)
results2 = cur.fetchall()
print("Subquery2: Display all movie names, Director, ratings, and gross collection where the genre is action")

# I am trying to print max 20 rows for display purpose. Displaying first 10 and last 10 rows for display purpose.

num_rows = len(results2)  # Determine the number of rows
max_rows = 20  # Determine the maximum number of rows to print

# Print the results
if num_rows > max_rows:
    # Print first 10 rows
    for row in results2[:10]:
        print(row)

    print("-"*100)
    print("-"*100)

    # Print last 10 rows
    for row in results2[-10:]:
        print(row)
else:
    # Print all results
    for row in results2:
        print(row)
```

```python
# Query: Display all details from both tables with the highest gross collection
query3 = """
    SELECT *
    FROM Table1
    INNER JOIN Table2 ON Table1.Movie_Name = Table2.Movie_Name
    WHERE Table2.Gross_Million_Dollars = (
        SELECT MAX(Gross_Million_Dollars) FROM Table2 WHERE Gross_Million_Dollars <> ''
    )
"""
cur.execute(query3)
results3 = cur.fetchall()
print("Subquery3: Display all details from both tables with the highest gross collection")

# I am trying to print max 20 rows for display purpose. Displaying first 10 and last 10 rows for display purpose.

num_rows = len(results3)  # Determine the number of rows
max_rows = 20  # Determine the maximum number of rows to print

# Print the results
if num_rows > max_rows:
    # Print first 10 rows
    for row in results3[:10]:
        print(row)

    print("-"*100)
    print("-"*100)

    # Print last 10 rows
    for row in results3[-10:]:
        print(row)
else:
    # Print all results
    for row in results3:
        print(row)
```

```python
# Query: Display all details from both tables with the highest ratings
query4 = """
    SELECT *
    FROM Table1
    INNER JOIN Table2 ON Table1.Movie_Name = Table2.Movie_Name
    WHERE Table1.Ratings = (
        SELECT MAX(Ratings) FROM Table1 WHERE Ratings <> ''
    )
"""
cur.execute(query4)
results4 = cur.fetchall()
print("Subquery4: Display all details from both tables with the highest ratings")

# I am trying to print max 20 rows for display purpose. Displaying first 10 and last 10 rows for display purpose.

num_rows = len(results4)  # Determine the number of rows
max_rows = 20  # Determine the maximum number of rows to print

# Print the results
if num_rows > max_rows:
    # Print first 10 rows
    for row in results4[:10]:
        print(row)

    print("-"*100)
    print("-"*100)

    # Print last 10 rows
    for row in results4[-10:]:
        print(row)
else:
    # Print all results
    for row in results4:
        print(row)
```

```python
# Query: Display all details from both tables with the lowest gross collection and lowest ratings

query5 = """
    SELECT *
    FROM Table1
    INNER JOIN Table2 ON Table1.Movie_Name = Table2.Movie_Name
    ORDER BY Table2.Gross_Million_Dollars ASC, Table1.Ratings ASC
    LIMIT 1
"""
cur.execute(query5)
results5 = cur.fetchall()
print("Subquery5: Display all details from both tables with the lowest gross collection and lowest ratings")

# I am trying to print max 20 rows for display purpose. Displaying first 10 and last 10 rows for display purpose.

num_rows = len(results5)  # Determine the number of rows
max_rows = 20  # Determine the maximum number of rows to print

# Print the results
if num_rows > max_rows:
    # Print first 10 rows
    for row in results5[:10]:
        print(row)

    print("-"*100)
    print("-"*100)

    # Print last 10 rows
    for row in results5[-10:]:
        print(row)
else:
    # Print all results
    for row in results5:
        print(row)

# Close the database dbection
db.close()
```
